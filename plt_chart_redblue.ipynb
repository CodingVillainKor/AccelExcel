```python
import openpyxl
import matplotlib.pyplot as plt
from openpyxl.drawing.image import Image

wb = openpyxl.load_workbook("data/Bias_correction_ucl.xlsx")
```

```python
sheet = wb["Sheet1"]
last_row = len(sheet["A"])

Fvalues_30up, Gvalues_30up = [], []
Fvalues_30down, Gvalues_30down = [], []
for i in range(2, last_row+1):
    if float(sheet["C{}".format(i)].value) > 30:
        Fvalues_30up.append(sheet["F{}".format(i)].value)
        Gvalues_30up.append(sheet["G{}".format(i)].value)
    else:
        Fvalues_30down.append(sheet["F{}".format(i)].value)
        Gvalues_30down.append(sheet["G{}".format(i)].value)

plt.scatter(Fvalues_30down, Gvalues_30down, s=0.2, color="red")
plt.scatter(Fvalues_30up, Gvalues_30up, s=0.2, color="blue")
plt.title("F and G")
plt.xlabel("x: F")
plt.ylabel("y: G")
plt.savefig("scatterchart.jpg")
```

```python
img = Image("scatterchart.jpg")
sheet.add_image(img, "I2")
wb.save("pyplot_chart_data_30.xlsx")
```
