```python
import openpyxl
import matplotlib.pyplot as plt
from openpyxl.drawing.image import Image

wb = openpyxl.load_workbook("data/Bias_correction_ucl.xlsx", data_only=True)
```

```python
sheet = wb["Sheet1"]
last_row = len(sheet["A"])

Fvalues = [sheet["F{}".format(i)].value for i in range(2, last_row+1)]
Gvalues = [sheet["G{}".format(i)].value for i in range(2, last_row+1)]

plt.scatter(Fvalues, Gvalues, s=0.2, color="blue")
plt.title("F and G")
plt.xlabel("x: F")
plt.ylabel("y: G")

plt.savefig("scatterchart.jpg")
img = Image("scatterchart.jpg")
sheet.add_image(img, "I2")
wb.save("pyplot_chart_data.xlsx")
```
